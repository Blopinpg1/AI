```python
from collections import deque

# State = (Farmer, Wolf, Goat, Cabbage)

def is_valid(state):
    F, W, G, C = state

    # Wolf alone with Goat
    if W == G and F != W:
        return False

    # Goat alone with Cabbage
    if G == C and F != G:
        return False

    return True


def get_next_states(state):
    F, W, G, C = state

    next_states = []

    # Farmer crosses alone
    new_state = (1 - F, W, G, C)
    if is_valid(new_state):
        next_states.append((new_state, "Farmer crosses alone"))

    # Farmer takes Wolf
    if F == W:
        new_state = (1 - F, 1 - W, G, C)
        if is_valid(new_state):
            next_states.append((new_state, "Farmer takes Wolf"))

    # Farmer takes Goat
    if F == G:
        new_state = (1 - F, W, 1 - G, C)
        if is_valid(new_state):
            next_states.append((new_state, "Farmer takes Goat"))

    # Farmer takes Cabbage
    if F == C:
        new_state = (1 - F, W, G, 1 - C)
        if is_valid(new_state):
            next_states.append((new_state, "Farmer takes Cabbage"))

    return next_states


def bfs():
    initial = (0, 0, 0, 0)
    goal = (1, 1, 1, 1)

    queue = deque([(initial, [])])
    visited = set()

    while queue:
        state, path = queue.popleft()

        if state == goal:
            return path

        if state in visited:
            continue

        visited.add(state)

        for next_state, action in get_next_states(state):
            if next_state not in visited:
                queue.append((next_state, path + [(action, next_state)]))

    return None


solution = bfs()

print("Solution:\n")

current = (0, 0, 0, 0)
print("Initial State:", current)

for i, (action, state) in enumerate(solution, 1):
    print(f"\nStep {i}")
    print("Action:", action)
    print("State :", state)

print("\nGoal Reached!")
```
