```python
from pprint import pprint 
n = 5
board = [['-']* n for i in range(n)]
pprint(board)
```

```python

# placing the obstacles
board[0][3] = 'X'
board[1][3] = 'X'
board[1][1] = 'X'
board[2][1] = 'X'
board[3][2] = 'X'
board[3][3] = 'X'
pprint(board)
```

```python
from collections import deque
def is_valid(board, new_state):
    row, col = new_state
    
    # rows = len(board)
    # cols = len(board[0])
    
    # board boundary check
    if row < 0 or   row > 4 or col < 0 or col > 4:
        return False
    
    # obstacle handling  
    if board[row][col] == 'X':
        return False
    
    return True
        
    
def get_next_states(board ,state):
    row ,col = state
    
    next_states = []
    
    #move up 
    new_state = (row -1 , col)
    if is_valid(board , new_state):
        next_states.append((new_state, "Move up"))
        
    #move down
    new_state = (row + 1 , col)
    if is_valid(board , new_state):
        next_states.append((new_state, "move down"))
        
    #move left
    new_state = (row , col - 1)
    if is_valid(board, new_state):
        next_states.append((new_state, "move left"))
        
        
    # move right
    new_state = (row , col+1)
    if is_valid(board , new_state):
        next_states.append((new_state, "move right"))
        
    return next_states


def bfs(board):
    initial = (0,0)
    goal = (4,4)
    queue = deque([(initial, [])])
    visited = set()
    
    
    while queue:
        state, path = queue.popleft()
        # print(f'      {state}     {path}')
        
        if state == goal:
            return path
        
        if state in visited:
            continue
        
        visited.add(state)
        
        for next_state , action in get_next_states(board,state):
            if next_state not in visited:
                queue.append((next_state , path + [(action , next_state)]))
                
    return None


solution = bfs(board)

print("Solution:\n")

current = (0, 0)
print("Initial State:", current)

for i, (action, state) in enumerate(solution, 1):
    print(f"\nStep {i}")
    print("Action:", action)
    print("State :", state)

print("\nGoal Reached!")

    
```
