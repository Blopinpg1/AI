```python
from collections import deque


# State = (Jug1 , Jug2)

def is_valid(state):
    J1 , J2 = state
    
    if   J1 < 0 or J1 > 4  or    J2 < 0 or J2 >3:
        return False
    
    return True
    


def get_next_states(state):
    J1 , J2 = state
    
    next_states = []
    
    # Jug one is filled 
    new_state = (4, J2)
    if is_valid(new_state):
        next_states.append((new_state, "Jug one (J1) is filled"))
    
    
    #Jug two is filled
    new_state = (J1, 3)
    if is_valid(new_state):
        next_states.append((new_state , "Jug two(J2) is filled"))

    #Jug one is emptied
    new_state = (0 , J2)
    if is_valid(new_state):
        next_states.append((new_state, "Jug one(J1) is emptied"))
     
    #Jug two is emptied   
    new_state = (J1 , 0)
    if is_valid(new_state):
        next_states.append((new_state , "Jug two(J2) is emptied"))
    
    # Pour from jug 1 to jug 2
    transfer = min(J1 , 3 - J2)
    new_state = (J1 - transfer , J2 + transfer)
    if is_valid(new_state):
        next_states.append((new_state, "Pour from J1 -> J2"))
    
    # Pour from jug 2 to jug 1
    transfer = min(4 - J1 , J2)
    new_state = (J1 + transfer , J2 - transfer)
    if is_valid(new_state):
        next_states.append((new_state, "Pour from J2 -> J1 "))


    return next_states


def is_goal(state):
    return state[0] == 2

    
        

def bfs():
    initial = (0,0)
    # goal = ((2,0),(2,1),(2,2),(2,3))
    queue = deque([(initial, [])])
    visited = set()
    
    
    while queue:
        state, path = queue.popleft()
        # print(f'      {state}     {path}')
        
        if is_goal(state):
            return path
        
        if state in visited:
            continue
        
        visited.add(state)
        
        for next_state , action in get_next_states(state):
            if next_state not in visited:
                queue.append((next_state , path + [(action , next_state)]))
                
    return None



solution = bfs()

print("Solution:\n")

current = (0, 0)
print("Initial State:", current)

for i, (action, state) in enumerate(solution, 1):
    print(f"\nStep {i}")
    print("Action:", action)
    print("State :", state)

print("\nGoal Reached!")



```
